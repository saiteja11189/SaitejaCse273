Data Frames in pandas

```python
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Marks": [85, 90, 88]
}
```

```python
print(data)
```

```python
type(data)
```

```python
marks = pd.DataFrame(data)
```

```python
print(marks)
```

```python
marks = pd.DataFrame(
    data,
    index = ["Student1", "Student2", "Student3"])
```

```python
print(marks)
```

```python
marks.set_index("Name", inplace=True)
```

```python
print(marks)
```
