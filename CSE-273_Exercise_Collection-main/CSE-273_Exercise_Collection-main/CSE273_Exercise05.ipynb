```python
import pandas as pd
import numpy as np

# Create DataFrame
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", None, "Eve", "Frank", "Grace"],
    "Department": ["IT", "HR", "IT", "Sales", "HR", "IT", "Sales"],
    "Role": ["Developer", "Manager", "Analyst", "Executive", "Analyst", "Developer", "Manager"],
    "Experience": [3, 10, 5, 2, 4, 8, 12],
    "Salary": [60000, 90000, 70000, 50000, 65000, 85000, 95000],
    "Score": [4.2, 4.8, 4.0, np.nan, 3.9, 4.5, np.nan]
})

```

```python
# Basic inspection
df.shape


```

```python
df.dtypes
```

```python
# Missing value analysis
df.isna().sum()

```

```python
# Employees per department
employees_per_dept = df.groupby("Department").size()

```

```python
# Average salary by department
avg_salary_dept = df.groupby("Department")["Salary"].mean()

```

```python
#Average salary by role
avg_salary_role = df.groupby("Role")["Salary"].mean()

```

```python
# Average performance score by department
avg_score_dept = df.groupby("Department")["Score"].mean()

```

```python
# Rows with missing names
missing_names = df[df["Name"].isna()]

```

```python
# New hires (missing score)
new_hires = df[df["Score"].isna()]

```

```python
# 1. Average (mean) Salary for each Department
df.groupby("Department")["Salary"].mean()

```

```python
# 2. Average (mean) Performance Score for each Role
df.groupby("Role")["Score"].mean()

```

```python
#3. Count of actual (non-null) Performance Score entries for each Department
df.groupby("Department")["Score"].count()

```

```python
# 4. Total number of employee records (rows) in each Department
df.groupby("Department").size()
```

```python
# Multiple Aggregations

# 1. Salary summary by Department (mean, max, min, count)
df.groupby("Department")["Salary"].agg(["mean", "max", "min", "count"])

```

```python
# 2. Performance Score summary by Role (mean, max, min, count)
df.groupby("Role")["Score"].agg(["mean", "max", "min", "count"])


```

```python
# Grouping by Multiple Columns

# 3. Average Salary grouped by Department and Role
df.groupby(["Department", "Role"])["Salary"].mean()

```

```python


# 4. Average Performance Score grouped by Department and Experience
df.groupby(["Department", "Experience"])["Score"].mean()
```
