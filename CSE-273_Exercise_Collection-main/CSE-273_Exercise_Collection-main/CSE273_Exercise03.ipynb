Task A1 — Create the Array Using Random Values

```python
import numpy as np

# Random electricity usage between 120 and 650
electricity = np.random.randint(120, 651, size=(4, 7))

print(electricity)
```

Task A2 — Array Attributes

```python
print("Shape:", electricity.shape)
print("Number of dimensions:", electricity.ndim)
print("Total elements:", electricity.size)
print("Data type:", electricity.dtype)
```

Task B1 — Hostel-Wise Metrics

```python
# Average usage per hostel (row-wise)
hostel_avg = np.mean(electricity, axis=1)
print("Average per hostel:", hostel_avg)

# Maximum usage per hostel
hostel_max = np.max(electricity, axis=1)
print("Maximum per hostel:", hostel_max)

# Hostel with highest average usage
highest_avg_hostel = np.argmax(hostel_avg) + 1
print("Hostel with highest average usage: H", highest_avg_hostel)
```

Task B2 — Day-Wise Metrics

```python
# Total usage per day (column-wise)
daily_total = np.sum(electricity, axis=0)
print("Total usage per day:", daily_total)

# Day with highest total consumption
highest_day = np.argmax(daily_total)
print("Day index with highest consumption:", highest_day)
```

Task C1 — Apply 20% Surcharge (Usage > 550) Using np.where

```python
penalized_usage = np.where(electricity > 550,
                           electricity * 1.20,
                           electricity)

print("Penalized usage:\n", penalized_usage)
```

Task D1 — Above/Below Hostel Weekly Average (Broadcasting)

```python
# Weekly average per hostel (shape: 4,)
weekly_avg = np.mean(electricity, axis=1)

# Broadcasting comparison to (4,7)
above_avg_matrix = electricity > weekly_avg[:, np.newaxis]

print("Above hostel average matrix:\n", above_avg_matrix)
```

Task E1 — Reshape to Single Row (1 × 28)

```python
single_row = electricity.reshape(1, 28)
print("Reshaped (1x28):\n", single_row)
```

Task E2 — Reshape to (2 × 14)

```python
reshaped_2x14 = electricity.reshape(2, 14)
print("Reshaped (2x14):\n", reshaped_2x14)
```

Task F1 — Copy & View

```python
ref_backup = electricity            # View (reference)
safe_backup = electricity.copy()   # Independent copy

print("Before modification:\n", electricity)

# Modify original
electricity[0, 0] = 999

print("After modification - Original:\n", electricity)
print("Reference Backup:\n", ref_backup)
print("Safe Backup:\n", safe_backup)
```
