```python
import numpy as np
import pandas as pd
```

```python

# Task 1.1: Create the DataFrame
data = {
    "Name": ["Aman", "Saurav", "Karan", "Shubham", "Abhishek", None, "Sameer"],
    "Age": [22, 23, np.nan, 22, 25, 24, 26],
    "Marks": [85, np.nan, 88, 90, 56, 72, np.nan],
    "Subject": ["Math", "Physics", "Math", "Chemistry", "Math", "Physics", "Chemistry"]
}


```

```python
df = pd.DataFrame(data)
```

```python

# Task 1.2: Basic Inspection

# 1. Shape of the dataset
df.shape
```

```python

# 2. Data types of columns
df.dtypes

```

```python
# 3. Statistical summary
df.describe()
```

```python
#Selection & Aggregation
df_name_marks = df[["Name", "Marks"]]
```

```python
df_marks_gt_80 = df[df["Marks"]>80]
```

```python
df_age_marks = df[(df["Age"]>22)&(df["Marks"]>80)]
```
