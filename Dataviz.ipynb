```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve","Hairish","Hemanth"],
    "Age": [25, 30, 35, 40, 45, 50, 55],
    "Gender" : ["Female", "Male", "Male", "Male", "Female", "Male", "Male"],
    "Marks": [85, 90, 95, 80, 88, 92, 96]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

Line plot for marks trend 

```python
plt.plot(df["Marks"])
plt.title("Marks of Students")
plt.xlabel("Student Index")
plt.ylabel("Marks")
plt.show()

```

Bar plot for comparision

```python
plt.bar(df["Name"], df["Marks"])
plt.title("Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.show()
```

Histogram for Disturbution of age

```python
plt.hist(df["Age"], bins=5)
plt.title("Age Distribution of Students")
plt.xlabel("Age")   
plt.ylabel("Frequency")
plt.show()
```

Scatter Plot

Relation b/w Age and Marks

```python
plt.scatter(df["Age"], df["Marks"])
plt.title("Age vs Marks")   
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()
```

pie Chart

Gender distribution

```python
gender_counts =df["Gender"].value_counts()
plt.pie(gender_counts,labels=gender_counts.index,autopct="%1.1f%%")
plt.title("Gender Distribution of Students")
plt.show()
```

Customizing Plots

Example with colors and grid

```python
plt.bar(data["Name"], data["Marks"],color=["blue","orange","green","red","purple","cyan","magenta"])
plt.title("Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

Sub Plots

show how graphs together

```python
plt.figure(figsize=(10,5))
plt.subplot(2,2,3)
plt.hist(df["Age"])
plt.title("Age Distribution")
plt.subplot(2,2,4)
plt.bar(df["Name"], df["Marks"])
plt.title("Marks by Student")
plt.show()
```
